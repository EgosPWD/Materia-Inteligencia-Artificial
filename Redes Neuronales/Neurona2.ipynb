```python
import numpy as np
from sklearn.neural_network import MLPClassifier



X = np.array([[0,0], [0,1], [1,0], [1,1]])
Y = np.array([0, 1, 1, 0]) # Darle la respuesta es entramiento de modelos supervisado

mlp = MLPClassifier( hidden_layer_sizes=(4,4), activation = "relu", max_iter=2000)
mlp.fit(X, Y)

print("Predicciones: ", mlp.predict(X))
print("Probabilidades: ", mlp.predict_proba(X))
```
