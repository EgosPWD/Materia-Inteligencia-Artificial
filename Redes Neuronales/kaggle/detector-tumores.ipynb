```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
```

```python
model = load_model('tumor_detector_model.h5')
img_path = 'test_image.jpg'
img = image.load_img(img_path, target_size=(224, 224))
img_array = image.img_to_array(img)
img_array = np.expand_dims(img_array, axis=0)
img_array = img_array / 255.0
```

```python
prediction = model.predict(img_array)
if prediction[0][0] > 0.5:
    print('Tumor detectado')
else:
    print('No se detectó tumor')
```

```python
def preprocess_image(img_path):
    img = image.load_img(img_path, target_size=(224, 224))
    img_array = image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)
    img_array = img_array / 255.0
    return img_array

def predict_tumor(model, img_array):
    prediction = model.predict(img_array)
    return prediction[0][0] > 0.5

img_path = 'test_image.jpg'
img_array = preprocess_image(img_path)
if predict_tumor(model, img_array):
    print('Tumor detectado')
else:
    print('No se detectó tumor')
```

```python
def show_image(img_path):
    img = cv2.imread(img_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(img)
    plt.axis('off')
    plt.show()

show_image('test_image.jpg')
```

```python
def segment_tumor(img_path):
    img = cv2.imread(img_path)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    _, thresh = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)
    return thresh
```

```python
segmented = segment_tumor('test_image.jpg')
plt.imshow(segmented, cmap='gray')
plt.axis('off')
plt.show()
```

```python
def draw_contours(img_path):
    img = cv2.imread(img_path)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    _, thresh = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)
    contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    cv2.drawContours(img, contours, -1, (0,255,0), 2)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(img)
    plt.axis('off')
    plt.show()

draw_contours('test_image.jpg')
```

```python
def process_folder(folder_path):
    for filename in os.listdir(folder_path):
        if filename.endswith('.jpg') or filename.endswith('.png'):
            img_path = os.path.join(folder_path, filename)
            img_array = preprocess_image(img_path)
            result = predict_tumor(model, img_array)
            print(f'{filename}: {"Tumor" if result else "No tumor"}')

process_folder('test_images_folder')
```
