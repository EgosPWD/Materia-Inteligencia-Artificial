```python
import numpy as np
from sklearn.linear_model import Perceptron
x = np.array([[0,0], [0,1], [1,0], [1,1]])
y = np.array([0,0,0,1])
modelo = Perceptron(max_iter=1000, tol = 1e-3)
modelo.fit(x,y)

print("Prediciones: ", modelo.predict(x))
print("pesos: ", modelo.coef_)
print("Dpias: " , modelo.intercept_)
```
