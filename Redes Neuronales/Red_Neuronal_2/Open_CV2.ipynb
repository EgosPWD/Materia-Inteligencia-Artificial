```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
from urllib.request import urlopen

```

```python
url = 'https://raw.githubusercontent.com/opencv/opencv/master/samples/data/lena.jpg'
resp = urlopen(url)
image = np.asarray(bytearray(resp.read()), dtype="uint8")
image = cv2.imdecode(image, cv2.IMREAD_COLOR)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

```

```python
y1, y2 = 100, 400
x1, x2 = 100, 400
start_y, start_x = int(y1 * 0.25), int(x1 * 0.25)
end_y, end_x = int(y2 * 0.75), int(x2 * 0.75)
cropped_image = image_rgb[start_y:end_y, start_x:end_x]
```

```python
new_width, new_height = 200, 200
resized_image = cv2.resize(image_rgb, (new_width, new_height))
```

```python
height, width = image_rgb.shape[:2]
center = (width // 2, height // 2)
angle = 45
rotation_matrix = cv2.getRotationMatrix2D(center, 45, 1)
rotated_image = cv2.warpAffine(image_rgb, rotation_matrix, (width, height))

```

```python
flipped_h = cv2.flip(image_rgb, 1)
flipped_v = cv2.flip(image_rgb, 0)
```

```python
plt.figure(figsize=(15, 10))

plt.subplot(2, 3, 1)
plt.title('Imagen Original')
plt.imshow(image_rgb)
plt.axis('off')

plt.subplot(2, 3, 2)
plt.title('Recortada')
plt.imshow(cropped_image)
plt.axis('off')

plt.subplot(2, 3, 3)
plt.title(f'Redimensionada ({new_width}x{new_height})')
plt.imshow(resized_image)
plt.axis('off')

plt.subplot(2, 3, 4)
plt.title(f'Rotada {angle}Â°')
plt.imshow(rotated_image)
plt.axis('off')

plt.subplot(2, 3, 5)
plt.title('Volteada Horizontalmente')
plt.imshow(flipped_h)
plt.axis('off')

plt.subplot(2, 3, 6)
plt.title('Volteada Verticalmente')
plt.imshow(flipped_v)
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python

print(f"Dimensiones de la imagen original: {image_rgb.shape}")
print(f"Dimensiones de la imagen recortada: {cropped_image.shape}")
print(f"Dimensiones de la imagen redimensionada: {resized_image.shape}")
```
